Bag of words and TF-IDF are the techniques use for feature extraction from text documents. Bag of words is a simple technique which converts text into a vector of numbers. It ignores grammar and order of words. TF-IDF is a weighted model commonly used for information retrieval problems. 

TF-IDF is a statistical measure that evaluates how relevant a word is to a document in a collection of documents. This is done by multiplying two metrics: how many times a word appears in a document, and the inverse document frequency of the word across a set of documents. TF-IDF is used for text classification and text clustering.

Corpus: A collection of text documents is called corpus. The plural form of corpus is corpora.

vocabularies: A vocabulary is a set of words known to a particular person, class, etc. or found in a particular text.

document: A document is a piece of text that may be classified in some category.

word: A word is a string of characters.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IMDB Dataset.csv')
df.head()
```

```python
doc1 = df['review'][0:100]
```

```python
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

df2 = cv.fit_transform(doc1)
```

```python
df2.shape
```

```python
print(cv.vocabulary_)
```

```python
print(df2.toarray())
```

```python
cv.transform(['This is a good movie']).toarray()
```

```python
df.head(100)
```

```python
cv.transform(['good movie']).toarray()
```

# TFIDF: Term Frequency - Inverse Document Frequency

```python
doc3 = df['review'][200:300]
doc3.head()
```

```python
doc3.shape
```

```python
#tfidf
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
df3 = tfidf.fit_transform(doc3).toarray()
```
